```python
# Dependencies
import numpy as np
import pandas as pd
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import datetime as dt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
Base.classes.keys()
first_row = session.query(Measurement).first()
first_row.__dict__
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract
from sqlalchemy import func
conn = engine.connect()
# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
select = '''
SELECT tobs, date  
FROM measurement
'''
june_temps_df = pd.read_sql(select, conn)
june_temps_df['date'] = pd.to_datetime(june_temps_df['date'])
june_temps_df = june_temps_df[june_temps_df['date'].dt.month == 6]
june_temps_df = june_temps_df.drop(columns=['date'])
june_temps_df
```

```python
#  3. Convert the June temperatures to a list.
june_temps = june_temps_df['tobs'].to_list()
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df= pd.DataFrame(june_temps, columns =['June temps'])
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_df.describe()
```

```python
select='''
SELECT tobs as Average_Temperature, date 
FROM Measurement 
WHERE date BETWEEN '2016-01-01' AND '2016-12-31'
'''

monthly_temps_df = pd.read_sql(select, conn)
monthly_temps_df['date'] = pd.to_datetime(monthly_temps_df['date'])
monthly_temps_df.groupby(pd.Grouper(key='date', freq='M')).mean()
```
